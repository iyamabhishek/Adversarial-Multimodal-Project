```python
import h5py
import numpy as np
```

```python
def partition_uni_modal():
    # Train
    with h5py.File('data/x_trimodal_train.h5','r') as h5f:
        x_t = np.array(h5f['data'][:,:,:300])
        x_a = np.array(h5f['data'][:,:,300:305])
        x_v = np.array(h5f['data'][:,:,305:])
    
    with h5py.File('data/x_unimodal_t_train.h5','w') as h5f:
        uni_modal = h5f.create_dataset('data', data = x_t)
    
    with h5py.File('data/x_unimodal_a_train.h5','w') as h5f:
        uni_modal = h5f.create_dataset('data', data = x_a)

    with h5py.File('data/x_unimodal_v_train.h5','w') as h5f:
        uni_modal = h5f.create_dataset('data', data = x_v)
    
    # Valid
    with h5py.File('data/x_trimodal_valid.h5','r') as h5f:
        x_t = np.array(h5f['data'][:,:,:300])
        x_a = np.array(h5f['data'][:,:,300:305])
        x_v = np.array(h5f['data'][:,:,305:])
    
    with h5py.File('data/x_unimodal_t_valid.h5','w') as h5f:
        uni_modal = h5f.create_dataset('data', data = x_t)
    
    with h5py.File('data/x_unimodal_a_valid.h5','w') as h5f:
        uni_modal = h5f.create_dataset('data', data = x_a)

    with h5py.File('data/x_unimodal_v_valid.h5','w') as h5f:
        uni_modal = h5f.create_dataset('data', data = x_v)
        
    # Test
    with h5py.File('data/x_trimodal_test.h5','r') as h5f:
        x_t = np.array(h5f['data'][:,:,:300])
        x_a = np.array(h5f['data'][:,:,300:305])
        x_v = np.array(h5f['data'][:,:,305:])
    
    with h5py.File('data/x_unimodal_t_test.h5','w') as h5f:
        uni_modal = h5f.create_dataset('data', data = x_t)
    
    with h5py.File('data/x_unimodal_a_test.h5','w') as h5f:
        uni_modal = h5f.create_dataset('data', data = x_a)

    with h5py.File('data/x_unimodal_v_test.h5','w') as h5f:
        uni_modal = h5f.create_dataset('data', data = x_v)
        
def partition_bi_modal():
    # Train
    with h5py.File('data/x_trimodal_train.h5','r') as h5f:
        x_t = np.array(h5f['data'][:,:,:300])
        x_a = np.array(h5f['data'][:,:,300:305])
        x_v = np.array(h5f['data'][:,:,305:])
    
    x_train_t_a = np.concatenate([x_t, x_a], axis = -1)
    x_train_a_v = np.concatenate([x_a, x_v], axis = -1)
    x_train_t_v = np.concatenate([x_t, x_v], axis = -1)

    with h5py.File('data/x_bimodal_t_a_train.h5','w') as h5f:
        bi_modal = h5f.create_dataset('data', data = x_train_t_a)
    
    with h5py.File('data/x_bimodal_a_v_train.h5','w') as h5f:
        bi_modal = h5f.create_dataset('data', data = x_train_a_v)

    with h5py.File('data/x_bimodal_t_v_train.h5','w') as h5f:
        bi_modal = h5f.create_dataset('data', data = x_train_t_v)
        
    # Valid
    with h5py.File('data/x_trimodal_valid.h5','r') as h5f:
        x_t = np.array(h5f['data'][:,:,:300])
        x_a = np.array(h5f['data'][:,:,300:305])
        x_v = np.array(h5f['data'][:,:,305:])
    
    x_train_t_a = np.concatenate([x_t, x_a], axis = -1)
    x_train_a_v = np.concatenate([x_a, x_v], axis = -1)
    x_train_t_v = np.concatenate([x_t, x_v], axis = -1)

    with h5py.File('data/x_bimodal_t_a_valid.h5','w') as h5f:
        bi_modal = h5f.create_dataset('data', data = x_train_t_a)
    
    with h5py.File('data/x_bimodal_a_v_valid.h5','w') as h5f:
        bi_modal = h5f.create_dataset('data', data = x_train_a_v)

    with h5py.File('data/x_bimodal_t_v_valid.h5','w') as h5f:
        bi_modal = h5f.create_dataset('data', data = x_train_t_v)
        
    # Test
    with h5py.File('data/x_trimodal_test.h5','r') as h5f:
        x_t = np.array(h5f['data'][:,:,:300])
        x_a = np.array(h5f['data'][:,:,300:305])
        x_v = np.array(h5f['data'][:,:,305:])
    
    x_train_t_a = np.concatenate([x_t, x_a], axis = -1)
    x_train_a_v = np.concatenate([x_a, x_v], axis = -1)
    x_train_t_v = np.concatenate([x_t, x_v], axis = -1)

    with h5py.File('data/x_bimodal_t_a_test.h5','w') as h5f:
        bi_modal = h5f.create_dataset('data', data = x_train_t_a)
    
    with h5py.File('data/x_bimodal_a_v_test.h5','w') as h5f:
        bi_modal = h5f.create_dataset('data', data = x_train_a_v)

    with h5py.File('data/x_bimodal_t_v_test.h5','w') as h5f:
        bi_modal = h5f.create_dataset('data', data = x_train_t_v)
```

```python
partition_uni_modal()
partition_bi_modal()
```
